```python
import pandas as pd
from sqlalchemy import create_engine
```

```python

df1 = pd.read_csv("./yelp_reviews_dashboard/db/2018_review_final_df.csv")
```

```python
df1.head()
```

```python
df1_cleaned = df1.loc[:,('stars', 'useful', 'funny', 'cool', 'text', 'date', 'polarity', 'subjectivity')]
```

```python
df1_cleaned.head()
```

```python

engine = create_engine("sqlite:///yelp_2018.db")
df1_cleaned.to_sql('review_2018',con=engine, if_exists="replace")
```

```python
engine.execute("SELECT * FROM review_2018").fetchall()
```
